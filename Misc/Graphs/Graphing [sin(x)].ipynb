```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(-0.001, 4*np.pi+0.001, np.pi/1000)
x_axis = np.arange(-0.001, 4*np.pi+0.001, np.pi/2)
f = np.tan(x)
```

```python
plt.plot(x, f)
plt.xticks(x_axis, ['0', 'π/2', 'π', '3π/2', '2π', '5π/2', '3π', '7π/2', '4π'])
plt.axhline(y=0, color='k')
plt.ylim(-10, 10)
```

```python
f2 = np.floor(f)

plt.scatter(x, f2, color='purple', s = 0.2)
plt.xticks(x_axis, ['0', 'π/2', 'π', '3π/2', '2π', '5π/2', '3π', '7π/2', '4π'])
plt.axhline(y=0, color='k', linewidth=0.3)
plt.ylim(-10, 10)
```
