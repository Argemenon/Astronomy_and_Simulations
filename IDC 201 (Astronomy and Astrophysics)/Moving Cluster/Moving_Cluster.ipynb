```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('D:\Personal Programs\IDC 201\Moving Cluster\hyades_data.txt')
data.columns = ['Name', 'RAH','RAM','RAS', 'DECH', 'DECM','DECS','PXX','RAP','DEP','VR']
```

```python
ra = (data.RAH + (data.RAM)/60 + (data.RAS)/3600)*15
dec = (data.DECH + (data.DECM)/60 + (data.DECS)/3600)
```

```python
#COM OF HYADES

com_ra = np.mean(ra)
com_dec = np.mean(dec)

print(com_ra, com_dec)
```

```python
# converting ra and dec of proper motion into degrees/year

rap = (data.RAP)/(3600*1000)

dep = (data.DEP)/(3600*1000)
```

```python
#total motion in 2,000,000 years:

del_ra = rap*2000000
del_dec = dep*2000000
```

```python
#Final position after 2 million years

ra_final = ra + del_ra
dec_final = dec + del_dec
```

```python
# calculation for mu

mu = np.sqrt((data.RAP)**2 + (data.DEP)**2)

distance = (data.VR*0.61*1000)/(4.74047*mu)

avg = np.mean(distance)
print(avg)
```

```python
plt.plot(distance)
plt.show
```

```python
# plot

plt.figure(figsize=(8,6))

for i in range(len(ra)):
    plt.plot([ra[i], ra_final[i]], [dec[i], dec_final[i]], color='red', linestyle='-', linewidth = 0.5)

plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.title('Change in Coordinates in 2 million years')

plt.grid(True, linestyle= '--', alpha= 0.7)

plt.show()
```
