```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import  style
```

```python
# Load the file with NumPy
data = np.genfromtxt(
    'moon.txt',
    dtype=str,  # Load as strings first
    delimiter=None,  # Use any whitespace as delimiter
    encoding='utf-8',
)

# Extract the first and last columns
dates = [" ".join(row[:2]) for row in data]  # Combine the first two columns (Month and Day)
angular_diameter = [float(row[-1]) for row in data]  # Last column converted to float

average = np.mean(angular_diameter)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(dates, angular_diameter, marker='.', label='Angular Diameter', color="blue")

# Highlight the last value
plt.axhline(
    y=average, color='r', linestyle='--', label=f'Average: {average:.2f}"'
)

# Find the index of the maximum value in y
max_index = np.argmax(angular_diameter)

# Get the x and y values at the maximum index
max_x = dates[max_index]
max_y = angular_diameter[max_index]

# Annotate the maximum value in the plot
plt.plot(max_x, max_y, 'ro')

# Find the index of the minimum value in y
min_index = np.argmin(angular_diameter)

# Get the x and y values at the minimum index
min_x = dates[min_index]
min_y = angular_diameter[min_index]

# Annotate the minimum value in the plot
plt.plot(min_x, min_y, 'go')

#annotation
plt.annotate(f"        {max_x},{max_y}", (max_x, max_y))

# Customize the x-axis labels
plt.xticks(rotation=45, fontsize=8)

#annotation
plt.annotate(f"            {min_x},{min_y}", (min_x, min_y))

# Customize the x-axis labels
plt.xticks(rotation=45, fontsize=8)

plt.xlabel("Date")
plt.ylabel("Angular Diameter")
plt.title("Moon Ephemerids - Angular Diameter Over Time")
plt.legend()
plt.tight_layout()  # Adjust layout for better spacing
plt.grid(True)
plt.show()
```
